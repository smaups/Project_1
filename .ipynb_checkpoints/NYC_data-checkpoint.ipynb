```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
```

```python
burrows = pd.read_csv("zip_borough.csv")
burrowszip = burrows.rename(columns={"zip":"Zipcode"})
```

```python
evictions_data = pd.read_csv("Evictions_clean_2.csv")
evictions_data.head()
evictions_data1 = evictions_data.rename(columns={'EVICTION_ZIP':"Zipcode"})
zip_eviction = evictions_data1.groupby('Zipcode').count()['EXECUTED_DATE']
zip_eviction.head()
```

```python
crime_data = pd.read_csv("Finalcrime1.csv")
crime_data.head()
zip_crime = crime_data.groupby('Zipcode').count()['ARREST_DATE']
zip_crime.to_frame().head()
```

```python
parks_data = pd.read_csv('Parks.csv')
parks1 = parks_data.rename(columns={'ZIPCODE':"Zipcode"})

zip_park = parks1.groupby('Zipcode').sum()['ACRES']
zip_park.to_frame().head()
```

```python
subway1_data = pd.read_csv("Subway.csv")
subway_data = subway1_data.rename(columns={'Line':"Subway Enterances"})
subway_data.head()
zip_subway = subway_data.groupby('Zipcode').count()['Subway Enterances']
zip_subway.to_frame().head()
```

```python
crime_subwaymerge = pd.merge(zip_crime, zip_subway, how="outer", on='Zipcode')
crime_subwaymerge.head()

Park_Crime_Subway = pd.merge(crime_subwaymerge, zip_park, how="outer", on='Zipcode')
Park_Crime_Subway.head()

totalmerge1 = pd.merge(Park_Crime_Subway, zip_eviction, how="outer", on="Zipcode")
totalmerge1['ARREST_DATE'] = totalmerge1['ARREST_DATE'].apply(pd.to_numeric)
totalmerge1.dropna().head()
totalmerge = totalmerge1.rename(columns={'ARREST_DATE':"Arrests","ACRES":"Acres of Parks","EXECUTED_DATE":"Evictions"})
```

```python
NYC_df1 = pd.merge(totalmerge, burrowszip, how='outer', on='Zipcode')
NYC_df = NYC_df1.rename(columns={"borough":"Borough"})
NYC_df.dropna().head()
```

```python
eviction = NYC_df.groupby(["Borough","Zipcode"]).mean()['Evictions']
evictionrate = (eviction/NYC_df["Evictions"].sum())*100
subway = NYC_df.groupby(["Borough", "Zipcode"]).mean()['Subway Enterances']
subwayrate = (subway/NYC_df['Subway Enterances'].sum())*100
park = NYC_df.groupby(["Borough", "Zipcode"]).mean()["Acres of Parks"]
parkrate = (park/NYC_df['Acres of Parks'].sum())*100
crime = NYC_df.groupby(["Borough", "Zipcode"]).mean()['Arrests']
```

```python
manhattan_subway = plt.scatter(evictionrate['Manhattan'], subwayrate['Manhattan'], marker='o', 
            facecolor='coral', s=crime*3,
            edgecolor='black', alpha=.75)
Bronx_subway = plt.scatter(evictionrate['Bronx'], subwayrate['Bronx'], marker='o', 
            facecolor='lightblue', s=crime*3,
            edgecolor='black', alpha=.75)
Staten_subway = plt.scatter(evictionrate['Staten'], subwayrate['Staten'], marker='o', 
            facecolor='green', s=crime*3,
            edgecolor='black', alpha=.75)
Queens_subway = plt.scatter(evictionrate['Queens'], subwayrate['Queens'], marker='o', 
            facecolor='purple', s=crime*3,
            edgecolor='black', alpha=.75)
Brooklyn_subway = plt.scatter(evictionrate['Brooklyn'], subwayrate['Brooklyn'], marker='o', 
            facecolor='red', s=crime*3,
            edgecolor='black', alpha=.75)

plt.text(3,2.5,"Circle Size Determined by Number of Crimes per Zipcode")
plt.xlabel("Evictions")
plt.ylabel("Subway Enterances")
plt.title("Number of Evictions and Subway Entrances in a Zipcode")
plt.grid()
subwaylegend = plt.legend([manhattan_subway, Bronx_subway, Staten_subway, Queens_subway, Brooklyn_subway],
                    ["Manhattan", "Bronx", "Staten", "Queens", "Brooklyn"], title="Boroughs", loc="best")

plt.show()
```

```python
crime_data = pd.read_csv("Finalcrime1.csv")
crime_data.head()
c_date = crime_data['ARREST_DATE'].str.split("/", n = 2, expand = True)
c_date
crime_data["Month"] = c_date[0]
crime_data["Date"] = c_date[1]
crime_data["Year"] = c_date[2]
crime_data["Month"] = crime_data["Month"].apply(pd.to_numeric)
lawcode = Counter(crime_data["LAW_CODE"])
```

```python
burrows = pd.read_csv("zip_borough.csv")
burrowszip = burrows.rename(columns={"zip":"Zipcode"})
crime_df = pd.merge(crime_data, burrowszip, how='inner', on="Zipcode")
eviction_df = pd.merge(evictions_data, burrowszip, how="inner", on="Zipcode")
eviction_df.head()
```

```python
bins = [0,1.1,2.1,3.1,4.1,5.1,6.1,7.1,8.1,9.1,10.1,11.1,12.1]
months_groups = ["January", "February", "March", 
          "April", "May", "June", 
          "July", "August", "September", 
          "October", "November", "December"]
```

```python
eviction_df['Months'] = pd.cut(eviction_df['Month'], bins, labels=months_groups)
number_evicted = eviction_df.groupby(['borough','Months'])['EXECUTED_DATE'].count()
```

```python
eviction_manhattan, = plt.plot(months_groups, number_evicted['Manhattan'], label="Manhattan", color='coral')
eviction_bronx, = plt.plot(months_groups, number_evicted['Bronx'], label='Bronx', color='lightblue')
# eviction_staten, = plt.plot(months_groups, number_evicted['Staten'], label='Staten', color='green')
eviction_queens, = plt.plot(months_groups, number_evicted['Queens'], label='Queens', color='purple')
eviction_brooklyn, = plt.plot(months_groups, number_evicted['Brooklyn'], label='Brooklyn', color='red')
plt.xticks(rotation=45)
plt.title("Number of Evictions in NYC By Month")
plt.xlabel("Month")
plt.ylabel("Evictions")
legend = plt.legend(handles=[eviction_manhattan,eviction_bronx,eviction_queens,eviction_brooklyn], loc="best")
ax = plt.gca().add_artist(legend)
# bronxlegend = plt.legend(handles=[eviction_bronx], loc="best")
# ax1 = plt.gca().add_artist(bronxlegend)
#                     ["Manhattan", "Bronx", "Staten", "Queens", "Brooklyn"], title="Boroughs", loc="best")
plt.grid()
```
