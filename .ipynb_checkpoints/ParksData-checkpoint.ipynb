```python
import pandas as pd
import numpy as np

parkrough_data = pd.read_csv("OpenData_ParkZones.csv")
park_df = pd.DataFrame(parkrough_data)
park_df.head()
```

```python
parkrough_data = park_df[['PROPNAME','ACRES','RETIRED','SUBCATEGOR','ZIPCODE','BOROUGH']]
parkrough_data.rename(columns={'PROPNAME':'Park Name',
                               'ACRES':'Acres',
                               'RETIRED':'Retired',
                               'SUBCATEGOR':'Sub Category',
                              'ZIPCODE':'Zipcode',
                              'BOROUGH':'Borough'})
parkrough_data.dtypes
```

```python
parkrough_data.loc[parkrough_data['PROPNAME'] == "Central Park",'ZIPCODE']
```

```python
parkrough_data.loc[parkrough_data['PROPNAME'] == "Central Park",'ACRES'].sum()
```

```python
parkrough_data.loc[parkrough_data['PROPNAME'] == "Central Park",:]
```

```python
def fix_zipcode(zipcode, idx):
    try:
        return zipcode.split(', ')[idx]
    except IndexError:
        return "None"
        

right_column = parkrough_data.loc[:,'ZIPCODE'].map(lambda x: fix_zipcode(x, 1))
right_column

left_column = parkrough_data.loc[:,'ZIPCODE'].map(lambda x: fix_zipcode(x, 0))
left_column
```

```python

parkrough_data['Zipcode1'] = left_column

parkrough_data['Zipcode2'] = right_column
```

```python
parkrough_data.loc[:,:]
```

```python
indexrange1 = range(1,591)
indexvalue1 = [y for y in indexrange1]

for row in parkrough_data:      
    parkrough_data['Index'] = indexvalue1
parkrough_data.set_index("Index")
```

```python
parkrough_data['ZIPCODE'] = parkrough_data['Zipcode1']
parkrough = parkrough_data[['PROPNAME','ACRES','ZIPCODE','RETIRED','SUBCATEGOR','BOROUGH',"Index"]]
parkrough
```

```python
parkzip1 = parkrough_data[['PROPNAME','ACRES','RETIRED','SUBCATEGOR','ZIPCODE','BOROUGH','Zipcode2']]

parkzip1 = parkzip1[parkzip1['Zipcode2'] != "None"]
len(parkzip1)
```

```python
indexrange = range(591,637)

indexvalue = [x for x in indexrange]
print(len(indexvalue))
# for row in parkzip1['Index']:
#     x = 700+ int(x)
#     x.appends(y)
for row in parkzip1:      
    parkzip1['Index'] = indexvalue
parkzip1.set_index("Index")
```

```python
parkzip1["ZIPCODE"] = parkzip1['Zipcode2']

parkzip = parkzip1[['PROPNAME','ACRES','ZIPCODE','RETIRED','SUBCATEGOR','BOROUGH',"Index"]]
parkzip
```

```python
parks_data = pd.concat((parkrough, parkzip), axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=False)
parks_data.set_index('Index')
```

```python
parks_data.to_csv(r'\Parks.csv')
```
