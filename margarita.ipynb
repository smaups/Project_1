```python
for index, row in subway_data.iterrows():
    subwaylat = row['Latitude']
    subwaylng = row['Longitude']
    subwayurl = (f"{url}latlng={subwaylat},{subwaylng}&key={key}")
    print(f"Processing record {index}")
#     print(f"Lat: {subwaylat}")
#     print(f"Lng: {subwaylng}")
    
    # Check if longitude is non-negative, move to the next row if so
    try:
        if subwaylng > 0:
            print(f"Index: {index}")
            print(f"Lng: {subwaylng}")
            continue
    except TypeError as e:
        print(f"subwaylng is non-numeric at index {index} with value {subwaylng}")
    response = requests.get(subwayurl).json() 
    output = response['results'][0]['address_components']    
    for item in output:
        #print(item['types'])
        for value in item['types']:
            if value == 'postal_code':
                subway_data.loc[index, "Zipcode"] = item['long_name']
```
